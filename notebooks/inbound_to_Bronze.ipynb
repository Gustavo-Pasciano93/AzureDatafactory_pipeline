Reading the data

```python
path = "dbfs:/mnt/data/inbound/dados_brutos_imoveis.json"

data = spark.read.json(path)
```

```python
display(data)
```

Dropping unnecessary columns

```python
data = data.drop("imagens")


data = data.drop("usuario")
```

```python
display(data)
```

Creating IDs

```python
from pyspark.sql import functions as F

from pyspark.sql.functions import col


```

```python
df_bronze = data.withColumn("id",col("anuncio.id"))
display(df_bronze)
```

Saving in the Bronze Layer

```python

path = "dbfs:/mnt/data/bronze/dataset_imoveis"
df_bronze.write.format("delta").mode("overwrite").save(path)
```
