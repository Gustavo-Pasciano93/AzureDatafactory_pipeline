Reading the Bronze Data

```python
path = "dbfs:/mnt/dados/bronze/dataset_imoveis/"
try:
    dbutils.fs.ls(path)
    df = spark.read.format("delta").load(path)
except Exception:
    df = None
```

```python
display(df)
```

Transforming the Json fields into Columns

```python
display(df.select("anuncio.*"))
```

```python
display(
    df.select("anuncio.*", "anuncio.endereco.*")
    )
```

```python
detailed_df =  df.select("anuncio.*", "anuncio.endereco.*")
```

```python
display(detailed_df)
```

Removing Unecessary columns

```python
df_silver = detailed_df.drop("caracteristicas")
```

```python
df_silver = df_silver.drop("endereco")
```

```python
display(df_silver)
```

Saving in the Silver Layer

```python
path = "dbfs:/mnt/data/silver/dataset_imoveis"
df_silver.write.format("delta").mode("overwrite").save(path)
```
